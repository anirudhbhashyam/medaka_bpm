```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('C:/Users/marci/Desktop/results_v2.CSV', sep=";")  
df
```

```python
loop1 = df.loc[df['Loop'] == 'LO001']
loop2 = df.loc[df['Loop'] == 'LO002']
loop2 = loop2['Heartrate (BPM)']
lopp2_list = loop2.tolist()


loop1.insert(1, "BPM loop2", lopp2_list)
loop1
df_new = loop1.rename(columns={'Heartrate (BPM)': 'BPM loop1'})
```

```python
df_new['average_loops'] = df_new[['BPM loop1', 'BPM loop2']].mean(axis=1)
df_new
```

```python
droped  = df_new.dropna()
droped
```

```python
plt.figure(figsize=(15,15))
sns.scatterplot(data=df_new, x='BPM loop1', y='BPM loop2'	)
```

```python
plt.figure(figsize=(15,15))

sns.histplot(data=df, x="Heartrate (BPM)")
```

```python
plt.figure(figsize=(15,15))

sns.histplot(data=df_new, x="average_loops")
```

```python
import numpy as np
def jitter(values,j):
    return values + np.random.normal(j,1.0,values.shape)

plt.figure(figsize=(15,15))
sns.scatterplot(data=droped, x=jitter(droped['BPM loop1'], 2), y='BPM loop2')
```
